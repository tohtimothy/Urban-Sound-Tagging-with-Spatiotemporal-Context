```python
import pandas as pd
df=pd.read_csv('annotations.csv')
df.head(60000)
```

```python
import datetime
from dateutil.relativedelta import relativedelta
```

```python
df["formatted_date"] = pd.to_datetime(df.year * 1000 + df.week * 10 + df.day * 1, format = '%Y%W%w')
df.head(60000)
```

```python
#storing of data with proper date and time to csv and xlsx file
df.to_csv('full_raw_data_date.csv', index=False)
df.to_excel('full_raw_data_date.xls', index=False)
```

```python
import pandas as pd
import random
import matplotlib.pyplot as plt
from datetime import datetime
from datetime import timedelta
from dateutil import rrule
```
