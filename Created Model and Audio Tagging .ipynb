```python
import numpy as np
from tqdm import tqdm
#iteration through every audio file and extract features
#using Mel=Frequency Cepstral Coefficients
extracted_features = []
for index_num,row in tqdm(metadata.iterrows()):
    file_name = os.path.join(os.path.abspath(audio_path),str(row["audio_filename"]))
    feature=row["Feature"]
    data=features_extractor(file_name)
    extracted_features.append([data,feature])
```

```python
#Conversion of extracted features to Pandas dataframe
extracted_features_df=pd.DataFrame(extracted_features,columns=['data','Feature'])
extracted_features_df.head()
```

```python
#Splitting dataset into independent and dependent dataset
X=np.array(extracted_features_df['data'].tolist())
y=np.array(extracted_features_df['Feature'].tolist())
```

```python
#When we need to work with test data we have to inverse transform
#Label Encoder
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
y=to_categorical(labelencoder.fit_transform(y))
```
